```python
x1= data["age"]                     # x1, x2 is  feature
x2= data["sallary"]
Target= data["sales"]             # Target
```

```python
import numpy as np
x1= np.array(x1)
x2-= np.array(x2)
Target= np.array(Target)
```

```python
# lets do data normalization         
x1= x1/34
x2= x2/113
```

```python
import random
w0= random.random()
w1= random.random()
w2= random.random()
k=0
lr= .01
while k<1000:

    y= w0 + w1*x1 + w2*x2
    yp= 1/ (1 + np.exp(-y))

    Loss= -1 * sum(Target * np.log(yp) + (1-Target) * np.log(1-yp))          
    print(Loss)

    w0= w0-lr * sum(yp-Target)
    w1= w1-lr * sum((yp-Target)*x1)
    w2= w2-lr * sum((yp-Target)*x2)

    k= k+1
```

```python
print(w0, w1, w2)
```
