```python
import pandas as pd
```

```python
data= pd.read_csv("customer_Leave_or_Stay.csv")
```

```python
data
```

```python
data= data.drop(["RowNumber","CustomerId","Surname"], axis= 1)
```

```python
data.head()
```

# Let's do the Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
k = LabelEncoder()

data["Geography"] = k.fit_transform(data["Geography"])

data["Gender"]= k.fit_transform(data["Gender"])
```

```python
data
```

# lets, Normalize the data

```python
data.max()
```

```python
data["CreditScore"] = data["CreditScore"]/850
```

```python
data["Balance"] = data["Balance"]/250898.09
```

```python
data["EstimatedSalary"] = data["EstimatedSalary"] / 199992.48
```

```python
data.head()
```

```python
y= data["Exited"]                              # Target
x=data.drop("Exited", axis= 1)                 # Featutes
```

```python
x.head(2)
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
x_train
```

```python
y_train
```

```python
def model(x , y):        # x, y is a parameter
    from sklearn.neighbors import KNeighborsClassifier
    neigh = KNeighborsClassifier(n_neighbors=3)
    neigh.fit(x , y) 
    
    from sklearn.naive_bayes import GaussianNB
    gnb = GaussianNB()
    gnb.fit(x , y)
    
    from sklearn.linear_model import LogisticRegression 
    logistic = LogisticRegression()
    logistic.fit(x , y)
    
    return neigh, gnb, logistic
```

```python
knn, nb, logistic = model(x_train , y_train)      # calling model function    x_train & y_train is argument
```

```python
def acc(knn,nb,logistic,x,y):
    a= knn.score(x,y)
    print("KNN acc is: ",a)
    
    b= nb.score(x,y)
    print("NB acc is: ",b)
    
    c= logistic.score(x,y)
    print("Logistic acc is: ",c)
    
    if a>b and a>c:
        print("\n KNN is Better")
        return knn
    
    elif b>c:
        print("\n NB is Better")
        return nb
    
    else:
        print("\n Logistic is Better")
        return logistic
    
```

```python
finalmodel= acc(knn, nb, logistic, x_test, y_test)
print(finalmodel)
```

```python
finalmodel.score(x_test,y_test)
```

# accuracy = 82%  approx

```python
finalmodel.score(x_train,y_train)
```

# model is ideal(bios-varience-trade-off)
