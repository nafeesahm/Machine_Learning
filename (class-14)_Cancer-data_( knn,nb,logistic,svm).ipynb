```python
import pandas as pd
```

```python
data= pd.read_csv("Cancer-data.csv")
```

```python
data.head()
```

```python
data= data.drop("id", axis= 1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
k = LabelEncoder()
data["diagnosis"] = k.fit_transform(data["diagnosis"])
```

```python
data
```

```python
data.max()
```

```python
# data Normalization

data["radius_mean"]= data["radius_mean"] / 28.11000
data["texture_mean"]= data["texture_mean"] / 39.28000
data["perimeter_mean"]= data["perimeter_mean"] / 188.50000
data["area_mean"]= data["area_mean"] / 2501.00000
data["perimeter_se"]= data["perimeter_se"] / 21.98000
data["area_se"]= data["area_se"] / 542.20000
data["radius_worst"]= data["radius_worst"] / 36.04000
data["texture_worst"]= data["texture_worst"] / 49.54000
data["perimeter_worst"]= data["perimeter_worst"] / 251.20000
data["area_worst"]= data["area_worst"] / 4254.00000
```

```python
data
```

```python
x= data.drop("diagnosis", axis= 1)              # x is feature
x
```

```python
y= data["diagnosis"]          #  y is  Target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
x_train
```

```python
y_train
```

```python
def model(x , y):
        from sklearn.neighbors import KNeighborsClassifier
        knn = KNeighborsClassifier(n_neighbors=3)
        knn.fit(x, y) 

        from sklearn.naive_bayes import GaussianNB
        nb = GaussianNB()
        nb.fit(x, y)

        from sklearn.linear_model import LogisticRegression 
        logistic = LogisticRegression()
        logistic.fit(x, y)
        
        from sklearn.svm import SVC                   # "Support vector classifier"
        svm = SVC(kernel='linear', C=1E10)            #  c is for avoid hard margin  , we increase c value for avoid 
        svm.fit(x, y)                                 #   c is a parameter in SVM

        return knn, nb, logistic, svm
```

```python
knn, nb, logistic, svm= model(x_train, y_train)              # model calling
```

```python
def acc(knn, nb, logistic, svm, x, y):
        k= knn.score(x, y)
        print("KNN score is: ",k)

        n= nb.score(x, y)
        print("NB score is : ",n)

        l= logistic.score(x, y)
        print("logistic score is: ",l)
        
        s= svm.score(x, y)
        print("svm score is: ",s)

        if k>n and k>l:
            if k>s:
                print("\n KNN is Better")
                return knn

        elif n>l and n>s:
            print("\n NB is Better")
            return nb
        
        elif l>s:
            print("\n logistic is better")
            return logistic
        
        else:
            print("\n SVM is better")
            return svm
```

```python
model= acc(knn, nb, logistic, svm, x_test, y_test)           # calling acc model
```

```python
model.score(x_test, y_test)
```

# accurecy =  93%

```python
model.score(x_train, y_train)
```
