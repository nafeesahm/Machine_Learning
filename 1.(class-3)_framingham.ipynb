```python
edu= [1.0,2.0,3.0,4.0]
for x in range(0,len(edu)):
    if x==1.0:
        data10= data.query('education==1.0')
        data10.to_csv("education_lebel_1.csv", index=0)
        
    elif x==2.0:
        data11= data.query('education==2.0')
        data11.to_csv("education_lebel_2.csv", index=0)
    
    elif x==3.0:
        data12= data.query('education==3.0')
        data12.to_csv("education_lebel_3.csv", index=0)
    
    else:
        data13= data.query('education==4.0')
        data13.to_csv("education_lebel_4.csv", index=0)
        
        
```
