```python
import numpy as np
x= np.array(x)
y= np.array(y)
```

```python
x,y
```

```python
w0= .1
w1= .2
w2= .3
lr= .000001
k= 0
while k<300:
    k= k+1
    yp = w0 + w1*x + w2*x**2
    Loss= 1/len(x) * sum((yp-y)**2)
    print(Loss)
    
    dLossdw0= 2/len(x) * sum((yp-y))
    dLossdw1= 2/len(x) * sum((yp-y)*x)
    dLossdw2= 2/len(x) * sum((yp-y)*x**2)
    
    w0= w0 - lr * dLossdw0
    w1= w1 - lr * dLossdw1
    w2= w2 - lr * dLossdw2
    
    #plt.scatter(x,y, color="Red")
    plt.plot(x,y, color="Red")
    plt.plot(x,yp)
```
